```python
from sklearn.linear_model import LinearRegression

# Prepare data for regression
X = df[['Item', 'Category', 'City', 'Month', 'Source']]
y = df['Price_per_Kg']

# Encode categories
categorical_features = ['Item', 'Category', 'City', 'Month', 'Source']
encoder = OneHotEncoder(handle_unknown='ignore')
```

```python
# Model Linear Regression
linear_model = Pipeline(
    steps=[
        ('encoder', ColumnTransformer(
            transformers=[('cat', encoder, categorical_features)],
            remainder='drop'
        )),
        ('regressor', LinearRegression())
    ]
)

linear_model.fit(X, y)
y_pred = linear_model.predict(X)

# Performance
mse = mean_squared_error(y, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y, y_pred)

print('Linear Regression Performance')
print('RMSE:', round(rmse, 2))
print('R:', round(r2, 3))
```

The linear regression model performs exceptionally well on the dataset, with an RMSE of 37.14 PKR per Kg and an R of 0.989. An R value this high indicates that the model explains 98.9% of the variation in food prices, meaning the encoded predictors, primarily Item and Category, capture almost all of the information needed to estimate price. The relatively low RMSE suggests that the models predictions deviate from actual prices by only about 37 PKR on average, which is small compared to the full price range in the dataset (roughly 40 to 1,200 PKR). This strong performance implies that food prices in this dataset are highly structured and predictable once categorical indicators are encoded, reflecting clear price differences across items and categories.

However, it is important to note that such a high R may also reflect overfitting, because the model is trained and evaluated on the same data. Since a linear regression with one-hot-encoded categories can closely approximate group averages, it often captures nearly all existing structure when no traintest split is applied. Nonetheless, the results indicate that price variation is strongly driven by categorical differences among items, and linear modeling is able to capture these relationships with high accuracy.

```python
# Model Random Forest Regression
rf_model = Pipeline(
    steps=[
        ('encoder', ColumnTransformer(
            transformers=[('cat', encoder, categorical_features)],
            remainder='drop'
        )),
        ('regressor', RandomForestRegressor(
            n_estimators=300,
            random_state=42,
            n_jobs=-1
        ))
    ]
)

rf_model.fit(X, y)
y_pred = rf_model.predict(X)

mse = mean_squared_error(y, y_pred)
rmse = mse ** 0.5
r2 = r2_score(y, y_pred)

print('Random Forest Regression Performance')
print('RMSE:', round(rmse, 2))
print('R:', round(r2, 3))
```

The Random Forest regression model shows outstanding performance, achieving an RMSE of 16.12 PKR per Kg and an R of 0.998, meaning it explains 99.8% of the variation in food prices. This represents a substantial improvement over the linear regression model, cutting prediction error by more than half. The extremely high R indicates that the model captures nearly all of the underlying patterns in the data, which is expected given the structured nature of food prices across items and categories. The low RMSE suggests that the models predictions deviate from true prices by only about 16 PKR on average, which is small relative to the overall price range.

These results demonstrate that Random Forest, a non-linear ensemble method, is particularly effective for this dataset because it can model complex interactions among categorical variables, such as differences across items, categories, and cities. However, such near-perfect performance also suggests the possibility of overfitting, especially because the model was evaluated on the same data it was trained on. Even so, the results strongly indicate that food prices in this dataset follow predictable and well-defined patterns that tree-based models can capture with very high accuracy.

```python
# Extract trained RandomForest model to see feature importance
rf = rf_model.named_steps['regressor']
ohe = rf_model.named_steps['encoder'].named_transformers_['cat']
feature_names = ohe.get_feature_names_out(categorical_features)

importances = pd.DataFrame({
    'feature': feature_names,
    'importance': rf.feature_importances_
}).sort_values(by='importance', ascending=False)

importances.head(20)
```

The feature importance results from the Random Forest model show which encoded variables contribute most to predicting food prices. The most influential predictor by far is Category_Meat, with an importance score of 0.353, indicating that simply knowing whether an item belongs to the meat category explains more than one-third of the models predictive power. This is consistent with the earlier findings that meat items, such as mutton, beef, chicken, and premium fish, form the highest-priced and most variable segment of the market.

Following category-level distinctions, several specific high-cost items also play a major role in determining price. Ghee, Cheese (Local), Spices (Mixed), and Tea (Loose) appear among the top contributors, each with substantial importance values. These items are consistently expensive across markets, so their encoded presence helps the model quickly identify price level ranges. The importance of these item-specific features suggests that individual commodities, especially those that are processed or imported, add significant predictive value beyond broader category labels.

Category-level indicators for beverages, vegetables, pulses, oil, grain, condiments, and dairy appear further down the list, each contributing smaller but still meaningful amounts. These categories hold lower price variability overall, so they provide less distinctive information to the model. The relatively small importance values for many items and categories reflect the fact that once the model accounts for the major price determinants, especially the meat category and a few high-price items, additional features add only minor refinements to the prediction.

Overall, the feature importance results confirm that food prices in this dataset are driven primarily by the type of product, with meat and other premium items exerting the strongest influence. The model relies most heavily on category distinctions and high-cost, high-variability items, illustrating that price behavior is highly structured and largely determined by commodity characteristics rather than city, month, or source indicators.

```python
# Cluster items into price tiers
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

item_prices = df.groupby('Item')['Price_per_Kg'].mean().to_frame()

scaler = StandardScaler()
item_prices_scaled = scaler.fit_transform(item_prices)

kmeans = KMeans(n_clusters=3, random_state=42)
item_prices['cluster'] = kmeans.fit_predict(item_prices_scaled)

item_prices.sort_values('cluster')
```

The item clustering results reveal three distinct price-based groups, each representing a different segment of Pakistans food market. Cluster 0 contains the largest collection of items and is characterized by low to moderately priced foods, including vegetables (spinach, tomato, onion, carrot), fruits (banana, oranges, apples, guava), grains (wheat, maize, flour), pulses (lentils and chickpeas), dairy items like yogurt and milk, and even some lower-priced oils. Prices in this cluster generally range from 40 to about 300 PKR per Kg, forming the core group of affordable everyday staples. These items are accessible to most households and contribute most heavily to regular food consumption patterns.

Cluster 1 represents the highest-priced premium items, which include mutton, pomfret fish, ghee, beef, local cheese, and loose tea. These items range roughly from 900 to over 1,200 PKR per Kg, placing them at the top of the market. Their elevated prices reflect factors such as higher production costs, lower supply, greater demand, and in some cases, import dependency. This cluster captures the commodities that exert the greatest financial pressure on consumers.

Cluster 2 forms a mid-priced segment, including items such as chicken (whole and breast), spices (mixed), fish (Rohu), and chicken breast. Prices in this group span from around 530 to 810 PKR per Kg, anchoring them between the low-cost staples and the premium goods. These foods are often essential but more expensive than the foundational staples, representing an important middle tier in consumer spending.

Together, the clusters reveal a clear three-tier structure in the food price landscape: affordable staples, moderately expensive essential items, and high-cost premium foods. This segmentation aligns closely with earlier findings on price distributions and highlights the consistent role of item type in determining food cost dynamics across Pakistans markets.

```python
# Cluster cities by their price profiles
city_matrix = df.pivot_table(
    values='Price_per_Kg',
    index='City',
    columns='Item',
    aggfunc='mean'
).fillna(0)

city_scaled = StandardScaler().fit_transform(city_matrix)

kmeans_cities = KMeans(n_clusters=4, random_state=42)
city_matrix['cluster'] = kmeans_cities.fit_predict(city_scaled)

city_matrix[['cluster']].sort_values('cluster')
```

```python
# Visualize city clusters
from sklearn.decomposition import PCA

# PCA
pca = PCA(n_components=2)
city_pca = pca.fit_transform(city_scaled)

city_matrix['PC1'] = city_pca[:, 0]
city_matrix['PC2'] = city_pca[:, 1]

# Plot PCA Scatterplot
plt.figure(figsize=(10, 7))
sns.scatterplot(
    data=city_matrix,
    x='PC1',
    y='PC2',
    hue='cluster',
    palette='deep',
    s=150
)

for city in city_matrix.index:
    plt.text(
        city_matrix.loc[city, 'PC1'] + 0.02,
        city_matrix.loc[city, 'PC2'] + 0.02,
        city,
        fontsize=9
    )

plt.title('City Clusters (PCA Visualization)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

The city clustering results, visualized through PCA, reveal four distinct groups of cities based on their overall food price patterns. Cluster 0, which includes Hyderabad, Quetta, and Sialkot, forms a group of relatively lower-priced cities. These cities appear close together in the PCA space, suggesting that their market structures and price levels are similar across a wide range of food items. Their placement on the right side of the plot reflects consistent affordability relative to other regions.

Cluster 1, composed of Karachi and Multan, represents cities with higher-than-average food prices. Karachi, as Pakistans largest metropolitan area, often exhibits elevated prices due to higher demand and urban cost pressures, while Multans placement suggests comparable pricing dynamics. These two cities cluster together at higher positions in the PCA space, reflecting their role as more expensive markets.

Cluster 2 includes Faisalabad, Islamabad, Lahore, and Rawalpindi, cities that fall into a mid-priced category. Their grouping indicates moderate price levels that are neither as low as Cluster 0 nor as high as Cluster 1. These cities are distributed fairly close to one another, suggesting similar patterns of pricing and availability, likely influenced by their developed urban infrastructure and balanced supply chains.

Finally, Cluster 3 consists solely of Peshawar, which stands out as a distinct outlier. Its isolated position at the top of the PCA plot shows that food prices in Peshawar differ from all other cities in the dataset. This uniqueness may reflect regional supply conditions, local market dynamics, or variations in transportation and import reliance.

Overall, the PCA visualization demonstrates that Pakistani cities naturally group into meaningful clusters based on food price behavior: a low-cost group, a high-cost group, a moderate-cost group, and one outlier. These clusters highlight how geographic and economic factors shape the structure of food markets across the country.

```python
city_matrix = df.pivot_table(
    values='Price_per_Kg',
    index='City',
    columns='Item',
    aggfunc='mean'
).fillna(0)

scaler = StandardScaler()
city_scaled = scaler.fit_transform(city_matrix)

kmeans_cities = KMeans(n_clusters=4, random_state=42)
city_matrix['cluster'] = kmeans_cities.fit_predict(city_scaled)

# Now merge cluster labels back to the main df
city_clusters = city_matrix[['cluster']].reset_index()   # City + cluster
df_clustered = df.merge(city_clusters, on='City', how='left')
```

```python
# Average price per Category within each cluster
radar_data = df_clustered.pivot_table(
    values='Price_per_Kg',
    index='cluster',
    columns='Category',
    aggfunc='mean'
)

# Normalize for nicer comparison (01 scale per category)
radar_norm = (radar_data - radar_data.min()) / (radar_data.max() - radar_data.min())
radar_norm
```

The normalized radar data summarizes how each city cluster differs in terms of average category-level food prices, scaled between 0 and 1 for easier comparison. Cluster 0, which includes the lowest-priced cities (Hyderabad, Quetta, Sialkot), shows very low normalized values for beverages, grain, oil, and vegetables but reaches the maximum for dairy, meat, and pulses. This indicates that although staple categories like grains and vegetables are extremely affordable in these cities, the relative prices of dairy, meat, and pulses are comparatively higher, but still low in absolute terms, because normalization reflects differences within categories, not overall price levels.
